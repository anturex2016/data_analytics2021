```python
import pandas as pd
import networkx as nx
import csv as csv
```

```python
df=pd.read_csv("data_analytics2021/datset/final_12.csv",encoding="ISO-8859-1")
df.head()
```

```python
co_graph=nx.from_pandas_edgelist(df,source="Author",target="Year")
```

```python
nx.info(co_graph)
```

```python
import matplotlib.pyplot as ptl
%matplotlib inline
```

```python
nx.draw(co_graph)
```

```python
nx.draw(co_graph,with_labels=True)
```

```python
ptl.figure(figsize=(20,20))
nx.draw_networkx(co_graph)
ptl.show()
```

```python
nx.degree_centrality(co_graph)
```

```python
nx.degree(co_graph)
```
