degree centrality measure

```python
sorted(nx.degree_centrality(co_graph).values())
```

```python
most_influential=nx.degree_centrality(co_graph)
```

```python
for i in sorted(most_influential,key=most_influential.get,reverse=True):
  print(i,most_influential[i])
```

# The most important connection

```python
most_important_link=nx.eigenvector_centrality(co_graph)
```

```python
for i in sorted(most_important_link,key=most_important_link.get,reverse=True):
  print(i,most_important_link[i])
```

```python
most_important_link1=nx.eigenvector_centrality(co_graph2)
for i in sorted(most_important_link1,key=most_important_link1.get,reverse=True):
  print(i,most_important_link1[i])
```

```python
most_important_link2=nx.eigenvector_centrality(co_graph1)
for i in sorted(most_important_link2,key=most_important_link2.get,reverse=True):
  print(i,most_important_link2[i])
```

# Closeness centrality
The shortest connection between nodes

```python
nx.shortest_path(co_graph,"Vitaly Osipov","Raymond Namyst")
```

# Betweeness centrality measure
Bridge/connection

```python
best_connector=nx.betweenness_centrality(co_graph)
for i in sorted(best_connector,key=best_connector.get,reverse=True):
  print(i,best_connector[i])
```

```python
df['author']=pd.factorize(df.Author)[0]
df['coauthor']=pd.factorize(df.CoAuthor)[0]


```

```python
df
```

```python
ptl.scatter(df['author'],df['coauthor'])
```

```python
from sklearn.cluster import KMeans

```

```python
k_rng= range(1,10)
sse=[]
for k in k_rng:
  km=KMeans(n_clusters=k)
  km.fit(df[['author','coauthor']])
  sse.append(km.inertia_)
```

```python
sse
```

```python
ptl.xlabel('K')
ptl.ylabel('sum of the suqared error')
ptl.plot(k_rng,sse)
```

```python
km=KMeans(n_clusters=3)
km
```

```python
y_predicted=km.fit_predict(df[['author','coauthor']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
df.info()
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]

ptl.scatter(df1.author,df1['coauthor'],color='green')
ptl.scatter(df2.author,df2['coauthor'],color='blue')
ptl.scatter(df3.author,df3['coauthor'],color='black')
ptl.xlabel('Author')
ptl.ylabel('Coauthor')
ptl.legend()
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
df[["author", "coauthor"]] = min_max_scaler.fit_transform(df[["author", "coauthor"]])
df
```

```python
km=KMeans(n_clusters=3)
y_predicted=km.fit_predict(df[["author", "coauthor"]])
y_predicted
```
